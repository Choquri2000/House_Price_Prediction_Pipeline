# Exploratory Data Analysis - House Price Prediction

## Key Findings from Analysis

### 1. Target Variable (SalePrice) Distribution
- **Skewness:** 1.88 (Highly positive skew)
- **Kurtosis:** 6.54 (Heavy tails, leptokurtic)
- **Action:** Log transformation (`np.log1p`) applied to normalize

### 2. Top Correlated Features with SalePrice
1. **OverallQual** (Overall Material and Finish Quality)
2. **GrLivArea** (Above grade living area square feet)
3. **TotalBsmtSF** (Total square feet of basement area)

### 3. Multicollinearity Detected
- `TotalBsmtSF` & `1stFlrSF` (high correlation)
- `GarageCars` & `GarageArea` (redundant info)
- **Action:** Kept stronger predictor, removed redundant features

### 4. Macroeconomic Variables Added
- GDP-Yr, HPI-Qrtr, LaborForce, Unemployment Rate, Interest Rate, Income metrics
- **Finding:** Low linear correlation (short 4-year dataset), but retained for potential non-linear relationships

### 5. Outliers Removed
- Two observations with extreme `GrLivArea` values removed (IDs: 1299, 524)

### 6. Missing Data Strategy
- Drop columns with >7 missing values
- Fill remaining with column means
